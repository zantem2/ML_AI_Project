```python
import pandas as pd
```

```python
df = pd.read_csv('SPY_Daily.csv')
df.head()
```

```python
#on close for each line run function
df['Act']='SELL'
df.head()
```

```python

stopLoss=0
actPrice=0
lastAct='SELL'
margin =0.03

def slact(dfRow):
    global lastAct, stopLoss, actPrice, margin
    price=dfRow.Close
    if lastAct == 'BUY':        
        if price < stopLoss:
            lastAct='SELL'
            stopLoss = price + (margin * price)
            actPrice=price
            print("SELL ","price",price, "SL ",stopLoss, "ACT ",actPrice)
        elif price > actPrice and price - (margin * price) > stopLoss:
            stopLoss = price - (margin * price)
            print("inc sl","price",price, "SL ",stopLoss, "ACT ",actPrice) 
    elif lastAct == 'SELL':
        if price > stopLoss:
            lastAct='BUY'
            print("BUY")
            stopLoss = price - (margin * price)
            actPrice=price
            print("BUY ","price",price, "SL ",stopLoss, "ACT ",actPrice)
        elif price < actPrice and price + (margin * price) < stopLoss:
            stopLoss = price + (margin * price)
            print("dec sl","price",price, "SL ",stopLoss, "ACT ",actPrice) 
    return lastAct
```

```python
"""
for each row of df
    out=act(Close)
    append out to a 1d array

"""
df['Act']=df.apply(slact,axis=1)
out=df['Act']
out.to_csv('Actions.csv')
```
